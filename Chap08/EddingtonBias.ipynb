# Draw a power law from 

```python
N=1000000
```

```python
u=np.random.uniform(low=0.00001,high=0.9999,size=N)
```

```python
alpha=-3
```

```python
data=np.power(u,1/alpha)
```

# Plot

```python
_, bins = np.histogram(np.log10(data + 1), bins='auto')
plt.hist(data,bins=10**bins,histtype='step')
plt.yscale('log')
plt.xscale('log')
```

```python
MIN=data.min()
MAX=data.max()
bins = 10 ** np.linspace(np.log10(MIN), np.log10(MAX), 50)
```

```python
plt.hist(data,bins=bins,histtype='step')
plt.yscale('log')
plt.xscale('log')
```

# add gaussian noise

```python
x=data
```

```python
y= np.random.normal(loc=0.0, scale=0.01, size=N)
```

```python
z=x+y
```

```python
z=np.delete(z,np.where(z<=0)[0])
```

```python
MIN=z.min()
MAX=z.max()
bins = 10 ** np.linspace(np.log10(MIN), np.log10(MAX), 50)
```

```python
plt.hist(z,bins=bins,histtype='step')
#plt.hist(x,bins=bins,histtype='step')
plt.yscale('symlog')
plt.xscale('symlog')
```
